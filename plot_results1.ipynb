```python
import torch
import matplotlib.pyplot as plt
import os
import numpy as np
import glob
```

```python
fols_add=glob.glob('./data/singles/add_seed=*/')
fols_a2b=glob.glob('./data/singles/a2b_seed=*/')
fols_both=glob.glob('./data/addp1/a2b_seed=*/')
```

```python
results_add=[]
results_a2b=[]
results_both=[]
for fols,results in zip([fols_add,fols_a2b,fols_both],[results_add,results_a2b,results_both]):
    for fol in fols:
        results.append(torch.load(fol+'results.pth'))
```

```python
plt.figure(figsize=(10,7))
cs={"add":'r',"a2b":'b',"both":'green'}
evals={"add":["add"],"a2b":["a2b"],"both":["add","a2b"]}
for name,results in zip(["add","a2b","both"],[results_add,results_a2b,results_both]):
    for result_add in results:
        save_steps=result_add['save_steps']
        #online_accs=result_add['online_accs']
        train_accs=result_add['train_accs']
        val_accs=result_add['val_accs']
        tr_accs_avg=[]
        val_accs_avg=[]
        for eval in evals[name]:
            tr_accs_avg.append(train_accs[eval])
            val_accs_avg.append(val_accs[eval])
        tr_accs_avg=np.array(tr_accs_avg).mean(axis=0)
        val_accs_avg=np.array(val_accs_avg).mean(axis=0)
        plt.plot(save_steps,tr_accs_avg,c=cs[name],ls="--")
        plt.plot(save_steps,val_accs_avg,c=cs[name],ls="-")
    plt.plot([],[],c=cs[name],label=name)
plt.plot([],[],c='black',ls="--",label="train")
plt.plot([],[],c='black',ls="-",label="val")
plt.legend()
plt.xscale('log')
plt.xlabel("Steps")
plt.ylabel("Accuracy")
```

```python
plt.figure(figsize=(10,7))
cs={"add":'r',"a2b":'b',"both":'green'}
evals=["add","a2b"]
for result_both in results_both:
    save_steps=result_both['save_steps']
    train_accs=result_both['train_accs']
    val_accs=result_both['val_accs']
    tr_accs_avg=[]
    val_accs_avg=[]
    for eval in evals:
        tr_accs_=train_accs[eval]
        val_accs_=val_accs[eval]
        plt.plot(save_steps,tr_accs_,c=cs[eval],ls="--")
        plt.plot(save_steps,val_accs_,c=cs[eval],ls="-")
        tr_accs_avg.append(tr_accs_)
        val_accs_avg.append(val_accs_)
    tr_accs_avg=np.array(tr_accs_avg).mean(axis=0)
    val_accs_avg=np.array(val_accs_avg).mean(axis=0)
    plt.plot(save_steps,tr_accs_avg,c=cs["both"],ls="--")
    plt.plot(save_steps,val_accs_avg,c=cs["both"],ls="-")
plt.plot([],[],c='r',label="add")
plt.plot([],[],c='b',label="a2b")
plt.plot([],[],c='green',label="both")
plt.plot([],[],c='black',ls="--",label="train")
plt.plot([],[],c='black',ls="-",label="val")
plt.legend()
plt.xscale('log')
plt.xlabel("Steps")
plt.ylabel("Accuracy")
```
